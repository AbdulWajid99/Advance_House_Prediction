FEATURE ENGINEERING

We will be performing all the below steps in Feature Engineering

1. Missing values
2. Temporal variables
3. Categorical variables: remove rare labels
4. Standarise the values of the variables to the same range

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

pd.pandas.set_option('display.max_columns',None)

df = pd.read_csv("E:\Data_Scientist\Python\Projects\Machine_Learning_Projects\Advance_House_Price_Prediction\App\Data\\train.csv")

df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,Y_train,x_test,y_test= train_test_split(df,df['SalePrice'],test_size=0.1,random_state=0) 
```

```python
#missing values Cat Features

catNull= [features for features in df.columns if df[features].isnull().sum()>1 and df[features].dtypes=='O'] # columns with null values

for features in catNull:
    print(features, np.round(df[features].isnull().mean(),4), "% is missing")
```

```python
def fill_nanValues(df,categ_nan):
    data=df.copy()
    data[categ_nan] = data[categ_nan].fillna('Missing')
    return data

newdf = fill_nanValues(df,catNull)

newdf[catNull].isnull().sum()
```

```python
newdf.head()
```

```python
# Numericals values with Null values

numNull  = [features for features in df.columns if df[features].isnull().sum()>1 and df[features].dtypes !='O']

for features in numNull:
    print(features, np.round(df[features].isnull().mean(),4), "% is missing")
```

```python
for features in numNull:
    #replace NAN with median since have outliers
    median_value = df[features].median()
    df[features+'nan'] = np.where(df[features].isnull(),1,0)
    df[features].fillna(median_value,inplace=True)

df[numNull].isnull().sum()
    
```

```python
## Temporal Variables (Date Time Variables)

tempFeatures = ['YearBuilt','YearRemodAdd','GarageYrBlt']

for feature in tempFeatures:
       
    df[feature]=df['YrSold']-df[feature]

df[tempFeatures].head()
```

```python
#Numerical Variables

import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df[feature]=np.log(df[feature])
```

 Handling Rare Categorical Feature
 
 We will remove categorical variables that are present less than 1% of the observations

```python
catFeatures = [features for features in df.columns if df[features].dtype =='O']

for features in catFeatures:
    temp = df.groupby(features)['SalePrice'].count()/len(df)
    tempdf = temp[temp>0.01].index
    df[features] = np.where(df[features].isin(tempdf),df[features],'Rare_var')
```

```python
df.head(50)
```

Feature Scaling

```python
for feature in catFeatures:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
#Select feature to scale,

feature_scale=[feature for feature in df.columns if feature not in ['Id','SalePrice']]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[feature_scale])
```

```python
# for train now
scaler.transform(df[feature_scale])
```

```python
# just concat id, salesprice since they didnt get fit tranfoemd
data = pd.concat([df[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(df[feature_scale]), columns=feature_scale)], #convert that output array to df again
                    axis=1)
```
